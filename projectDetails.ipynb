### User Requirements / Features (sorted by importance)

1. **Authentication & Account**
	- Login / Logout
	- User registration

2. **Browse & Discover Shops**
	- View barber shops and details (services, prices, hours)
	- Search nearby barber shops
	- View shop location on map

3. **Appointments & Queue**
	- Book and cancel appointments
	- View appointment history
	- View queue status and get estimated wait time

4. **Notifications**
	- Receive notifications for upcoming appointments

5. **Reviews**
	- Rate and review barber shops


### Shop Owner Requirements / Features (sorted by importance)

1. **Authentication & Registration**
	- Shop owner login / logout
	- Shop registration

2. **Appointment Management**
	- View, confirm, and cancel appointments
	- Pause / resume accepting appointments
	- Receive notifications for new appointments

3. **Shop Profile & Services**
	- Manage shop profile: services, prices, hours

4. **Queue & Customer Interaction**
	- Manage queue status
	- View and respond to customer reviews and ratings
	- View shop location on map

5. **Reporting & History**
	- View appointment history


### Database Schemas (core tables)

- **users**
  - `user_id` (PK)
  - `username`
  - `email`
  - `password_hash`
  - `phone_number`
  - `created_at`, `updated_at`

- **shops**
  - `shop_id` (PK)
  - `owner_id` (FK -> users.user_id)
  - `shop_name`, `address`, `phone_number`, `email`
  - `opening_time`, `closing_time`
  - `rating`

- **appointments**
  - `appointment_id` (PK)
  - `user_id` (FK -> users.user_id)
  - `shop_id` (FK -> shops.shop_id)
  - `appointment_time`
  - `status` (booked | confirmed | canceled)
  - `created_at`, `updated_at`

- **queues**
  - `queue_id` (PK)
  - `shop_id` (FK -> shops.shop_id)
  - `current_queue_length`
  - `estimated_wait_time`
  - `updated_at`

- **reviews**
  - `review_id` (PK)
  - `user_id` (FK -> users.user_id)
  - `shop_id` (FK -> shops.shop_id)
  - `rating`, `comment`
  - `created_at`, `updated_at`

- **notifications**
  - `notification_id` (PK)
  - `user_id` (FK -> users.user_id)
  - `message`, `is_read`
  - `created_at`, `updated_at`

### API Endpoints (grouped)

**User endpoints**
- `POST /api/register` — Register a new user
- `POST /api/login` — Authenticate user and return token
- `POST /api/logout` — Invalidate user token
- `GET /api/shops` — List all barber shops
- `GET /api/shops/search` — Search nearby barber shops (by location)
- `POST /api/appointments` — Book an appointment
- `DELETE /api/appointments/{id}` — Cancel an appointment
- `GET /api/appointments/history` — Get user's appointment history
- `GET /api/notifications` — Get user's notifications
- `POST /api/reviews` — Submit a review for a barber shop

**Shop owner endpoints**
- `POST /api/shops/register` — Register a new shop
- `POST /api/shops/login` — Authenticate shop owner and return token
- `POST /api/shops/logout` — Invalidate shop owner token
- `GET /api/shops/{id}` — Get shop details
- `PUT /api/shops/{id}` — Update shop profile
- `GET /api/appointments` — Get all appointments for the shop
- `POST /api/appointments/{id}/confirm` — Confirm an appointment
- `POST /api/appointments/{id}/cancel` — Cancel an appointment
- `GET /api/reviews` — Get all reviews for the shop
- `POST /api/reviews/{id}/respond` — Respond to a customer review
